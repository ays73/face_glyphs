```python
#imports
from transformers import pipeline
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)
import nltk
nltk.download('punkt_tab')
from nltk import tokenize
import kagglehub
import pandas as pd
import os
```

```python
path = kagglehub.dataset_download("lakshmi25npathi/imdb-dataset-of-50k-movie-reviews") #download imdb dataset
```

```python
imdb = pd.read_csv(os.path.join(path,'IMDB Dataset.csv'))
```

```python
imdb_short = imdb.sort_values(by = "review", key=lambda x: x.str.len()) #sort by length to get shorter reviews
```

```python
def face_glyph(sentence):
  """
  creates face glyph for one sentence
  """
  classification = classifier(sentence)
  emotions = []
  for i in range(7):
    emotions.append(list(classification[0][i].values()))
  emotion, value = max(emotions, key=lambda x: x[1])
  if emotion == "joy":
    emotion = "happiness"
  emotion_command = "\\" + emotion + "{{{}}}".format(round(value,3))
  emotion_command = "{{{}}}".format(emotion_command)
  if emotion == "neutral":
    emotion_command = "{\\neutral}"
  return("\\faceglyph{0}{0}{0}{0}{0}{0}"+ emotion_command)

def sentence_glyphs(sentences):
  """
  creates face glyphs for multiple sentences
  """
  sentences_split = tokenize.sent_tokenize(sentences)
  out = ""
  for sentence in sentences_split:
    out = out + sentence[:-1] + " " + face_glyph(sentence) + sentence[-1] + "\n"
  print("Copy and paste the following command into Latex: \n")
  print(out)

def get_review(idx):
  return(imdb_short.iloc[idx]["review"])

def review_glyph(idx):
  return(sentence_glyphs(get_review(idx)))
```
